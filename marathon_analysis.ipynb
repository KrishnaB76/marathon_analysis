```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df= pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
# We want to filter our data for USA only,event distance of 50km or 50 mi and races which took place in 2020
#this will filter out races with 50km and 50mi
df[df["Event distance/length"].isin(['50km','50mi'])]
df.head(3)
```

```python
#this will filter out races which took place in 2020 only
df[df["Year of event"] == 2020]
df.head(3)
```

```python
#filtering out races whic took place in USA only
df[df["Event name"].str.split('(').str.get(1).str.split(")").str.get(0)=="USA"]
df.head(3)
```

```python
#combining all of the filters together 
df2=df[(df["Event distance/length"].isin(['50km','50mi']))&(df["Year of event"] == 2020)&(df["Event name"].str.split('(').str.get(1).str.split(")").str.get(0)=="USA")]
```

```python
#Cleaning the data
#to remove USA from event name because whole data is of USA
# Creating a new column for athlete age and removing h from Athlete performance
df2["Athlete age"]= 2020- df2["Athlete year of birth"]
df2["Athlete performance"] = df2["Athlete performance"].str.split(" ").str.get(0)
df2["Event name"]=df2["Event name"].str.split("(").str.get(0)
```

```python
#dropping the columns which is not important
df2 = df2.drop(["Athlete club","Athlete country","Athlete age category","Athlete year of birth"],axis=1)
```

```python
#checking null values in the dataframe
df2.isna().sum()
```

```python
#check whether dataframe contains duplicate values
df2[df2.duplicated()==True]
```

```python
#droppping the null values in the dataframe
df2=df2.dropna()
#reseting the index numbers
df2=df2.reset_index(drop=True)
df2.head(3)
```

```python
#fix the data types
df2["Athlete age"] =df2["Athlete age"].astype(int)
df2["Athlete average speed"] =df2["Athlete average speed"].astype(float)
df2.dtypes
```

```python
#renaming column names to avoid mistakes in the future
df2 =df2.rename(columns={"Year of event":"year_of_event",
"Event dates":"race_day",                  
"Event name":"race_name",                    
"Event distance/length":"race_length",        
"Event number of finishers" :"race_number_of_finishers",    
"Athlete performance":"athlete_performance",           
"Athlete gender":"athlete_gender",              
"Athlete average speed":"athlete_average_speed",     
"Athlete ID":"athlete_id",                     
"Athlete age":"athlete_age"})
df2.head(3)
```

```python
#reordering the columns
df3=df2[["race_day","race_name","race_length","athlete_id","athlete_gender","athlete_age","athlete_performance","athlete_average_speed","race_number_of_finishers","athlete_average_speed",]]
```
